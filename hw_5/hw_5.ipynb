# PR

Данных в лекции недостаточно, чтобы понять, как ее строить в C++. Будем перестраивать модель и генерировать данные здесь автоматически с помощью scikitlearn

```python
import random
```

```python
data = []
for i in range(10000):
    x = random.random()
    y = random.random()
    d = [max([x, y]), 1]
    data.append(d)

    x = random.random()
    y = random.random()
    d = [1 - max([x, y]), 0]
    data.append(d)
```

```python
data
```

```python
x = []
for i in data:
    x.append(i[0])
y = []
for i in data:
    y.append(i[1])
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
xx = np.asarray(x)
yy = np.asarray(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(xx.reshape(-1, 1) , yy, test_size=0.3, shuffle=True, random_state=0xEDA)
lr = LogisticRegression(max_iter=100)
lr.fit(X_train, y_train)
```

```python
predictions = lr.predict_proba(X_test)
precision, recall, _ = metrics.precision_recall_curve(y_test, predictions[:,1])
disp = metrics.PrecisionRecallDisplay(precision=precision, recall=recall)
disp.plot()
```
